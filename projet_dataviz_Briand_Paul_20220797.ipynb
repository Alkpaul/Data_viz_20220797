```python
# Preprocessing des données
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Configuration pour l'affichage
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

print("=== PREPROCESSING DES DONNÉES ===")
print(f"Dataset initial : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

```python
# Fonction pour nettoyer les valeurs de balance
def clean_balance_value(value):
    """
    Convertit une valeur de balance en float
    Gère les cas : '-', valeurs avec espaces, virgules comme séparateurs décimaux
    """
    if pd.isna(value) or value == '-' or value.strip() == '':
        return np.nan
    
    # Nettoyer la chaîne : supprimer les espaces et remplacer les virgules par des points
    cleaned = str(value).strip().replace(' ', '').replace(',', '.')
    
    try:
        return float(cleaned)
    except ValueError:
        return np.nan

# Appliquer le nettoyage aux colonnes de balance
balance_columns = [col for col in df.columns if 'Balance Sortie' in col]
print(f"Colonnes de balance identifiées : {len(balance_columns)}")
print("Colonnes :", balance_columns)

# Créer une copie pour le preprocessing
df_clean = df.copy()

# Nettoyer toutes les colonnes de balance
for col in balance_columns:
    df_clean[col] = df_clean[col].apply(clean_balance_value)

print(f"\nNettoyage terminé. Types de données après nettoyage :")
for col in balance_columns:
    print(f"  {col}: {df_clean[col].dtype}")
```

```python
# Vérification du nettoyage
print("=== VÉRIFICATION DU NETTOYAGE ===")

# Afficher quelques exemples avant/après
print("\nExemples de transformation :")
sample_data = df[balance_columns[:3]].head(3)
print("AVANT nettoyage :")
print(sample_data)

print("\nAPRÈS nettoyage :")
sample_clean = df_clean[balance_columns[:3]].head(3)
print(sample_clean)

# Statistiques sur les valeurs nulles
print(f"\n=== STATISTIQUES SUR LES VALEURS NULLES ===")
for col in balance_columns:
    null_count = df_clean[col].isna().sum()
    total_count = len(df_clean)
    null_percentage = (null_count / total_count) * 100
    print(f"{col}: {null_count:,} valeurs nulles ({null_percentage:.1f}%)")

# Statistiques descriptives sur les colonnes nettoyées
print(f"\n=== STATISTIQUES DESCRIPTIVES ===")
print("Statistiques sur les colonnes de balance (valeurs non-nulles) :")
for col in balance_columns:
    non_null_data = df_clean[col].dropna()
    if len(non_null_data) > 0:
        print(f"\n{col}:")
        print(f"  Nombre de valeurs non-nulles: {len(non_null_data):,}")
        print(f"  Min: {non_null_data.min():,.2f}")
        print(f"  Max: {non_null_data.max():,.2f}")
        print(f"  Moyenne: {non_null_data.mean():,.2f}")
        print(f"  Médiane: {non_null_data.median():,.2f}")
```

```python
# Restructuration des données pour faciliter l'analyse
print("=== RESTRUCTURATION DES DONNÉES ===")

# Créer un DataFrame "long format" pour les analyses temporelles
df_melted = pd.melt(
    df_clean, 
    id_vars=['Postes', 'Sous-postes', 'Indicateurs de synthèse', 'Indicateurs de détail', 
             'Compte', 'Nature Budgétaire', 'Programme', 'Libellé Ministère'],
    value_vars=balance_columns,
    var_name='Année',
    value_name='Balance'
)

# Extraire l'année du nom de la colonne
df_melted['Année'] = df_melted['Année'].str.extract(r'(\d{4})').astype(int)

print(f"Dataset restructuré : {df_melted.shape[0]:,} lignes")
print(f"Période couverte : {df_melted['Année'].min()} - {df_melted['Année'].max()}")
print(f"Nombre d'années : {df_melted['Année'].nunique()}")

# Afficher un échantillon du dataset restructuré
print("\nÉchantillon du dataset restructuré :")
print(df_melted.head(10))
```

```python
# Exploration des dimensions du dataset
print("=== EXPLORATION DES DIMENSIONS ===")

# Analyser les différentes dimensions
dimensions = {
    'Postes': df_melted['Postes'].nunique(),
    'Sous-postes': df_melted['Sous-postes'].nunique(), 
    'Indicateurs de synthèse': df_melted['Indicateurs de synthèse'].nunique(),
    'Indicateurs de détail': df_melted['Indicateurs de détail'].nunique(),
    'Ministères': df_melted['Libellé Ministère'].nunique()
}

print("Nombre de valeurs uniques par dimension :")
for dim, count in dimensions.items():
    print(f"  {dim}: {count:,}")

# Afficher les principales catégories
print(f"\n=== PRINCIPALES CATÉGORIES ===")

print("\nTop 10 des Postes :")
top_postes = df_melted['Postes'].value_counts().head(10)
for poste, count in top_postes.items():
    print(f"  {poste}: {count:,} entrées")

print("\nTop 10 des Ministères :")
top_ministres = df_melted['Libellé Ministère'].value_counts().head(10)
for ministere, count in top_ministres.items():
    print(f"  {ministere}: {count:,} entrées")

print("\nTop 10 des Indicateurs de synthèse :")
top_indicateurs = df_melted['Indicateurs de synthèse'].value_counts().head(10)
for indicateur, count in top_indicateurs.items():
    print(f"  {indicateur}: {count:,} entrées")
```

```python
# VISUALISATION 1: Évolution temporelle des balances totales
print("=== VISUALISATION 1: ÉVOLUTION TEMPORELLE ===")

# Calculer les totaux par année (en excluant les valeurs nulles)
balance_totals = df_melted.groupby('Année')['Balance'].sum()
balance_totals_millions = balance_totals / 1e6  # Convertir en millions d'euros

# Créer le graphique
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Graphique 1: Évolution des totaux
ax1.plot(balance_totals_millions.index, balance_totals_millions.values, 
         marker='o', linewidth=3, markersize=8, color='#2E86AB')
ax1.set_title('Évolution des Balances Totales de l\'État (2012-2022)', fontsize=16, fontweight='bold')
ax1.set_xlabel('Année', fontsize=12)
ax1.set_ylabel('Balance (Millions d\'euros)', fontsize=12)
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='Équilibre budgétaire')

# Ajouter des annotations pour les points remarquables
for year, value in balance_totals_millions.items():
    if abs(value) > 50000:  # Annoter les valeurs importantes
        ax1.annotate(f'{value:,.0f}M€', (year, value), 
                    textcoords="offset points", xytext=(0,10), ha='center',
                    fontsize=9, fontweight='bold')

ax1.legend()
ax1.set_xticks(balance_totals_millions.index)

# Graphique 2: Distribution des balances par année
balance_data_by_year = [df_melted[df_melted['Année'] == year]['Balance'].dropna() 
                       for year in sorted(df_melted['Année'].unique())]

ax2.boxplot(balance_data_by_year, labels=sorted(df_melted['Année'].unique()))
ax2.set_title('Distribution des Balances par Année', fontsize=14, fontweight='bold')
ax2.set_xlabel('Année', fontsize=12)
ax2.set_ylabel('Balance (Euros)', fontsize=12)
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Afficher les statistiques
print(f"\nStatistiques des balances totales par année :")
for year in sorted(balance_totals_millions.index):
    total = balance_totals_millions[year]
    print(f"  {year}: {total:,.0f} millions d'euros")
```

```python
# VISUALISATION 2: Analyse par ministères
print("=== VISUALISATION 2: ANALYSE PAR MINISTÈRES ===")

# Calculer les totaux par ministère (sur toute la période)
ministere_totals = df_melted.groupby('Libellé Ministère')['Balance'].sum().sort_values(ascending=False)
ministere_totals_millions = ministere_totals / 1e6

# Créer le graphique
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Graphique 1: Top 10 des ministères par balance totale
top_10_ministres = ministere_totals_millions.head(10)
colors = plt.cm.Set3(np.linspace(0, 1, len(top_10_ministres)))

bars1 = ax1.barh(range(len(top_10_ministres)), top_10_ministres.values, color=colors)
ax1.set_yticks(range(len(top_10_ministres)))
ax1.set_yticklabels([label[:30] + '...' if len(label) > 30 else label for label in top_10_ministres.index])
ax1.set_xlabel('Balance Totale (Millions d\'euros)', fontsize=12)
ax1.set_title('Top 10 des Ministères par Balance Totale (2012-2022)', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='x')

# Ajouter les valeurs sur les barres
for i, (bar, value) in enumerate(zip(bars1, top_10_ministres.values)):
    ax1.text(value + (max(top_10_ministres.values) * 0.01), i, f'{value:,.0f}M€', 
             va='center', fontsize=9, fontweight='bold')

# Graphique 2: Évolution des 5 principaux ministères
top_5_ministres = ministere_totals_millions.head(5).index
ministere_evolution = df_melted[df_melted['Libellé Ministère'].isin(top_5_ministres)].groupby(['Année', 'Libellé Ministère'])['Balance'].sum().unstack(fill_value=0)
ministere_evolution_millions = ministere_evolution / 1e6

for ministere in top_5_ministres:
    if ministere in ministere_evolution_millions.columns:
        ax2.plot(ministere_evolution_millions.index, ministere_evolution_millions[ministere], 
                marker='o', linewidth=2, label=ministere[:25] + '...' if len(ministere) > 25 else ministere)

ax2.set_title('Évolution des 5 Principaux Ministères (2012-2022)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Année', fontsize=12)
ax2.set_ylabel('Balance (Millions d\'euros)', fontsize=12)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='red', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Statistiques détaillées
print(f"\nStatistiques par ministère :")
for ministere, total in top_10_ministres.items():
    print(f"  {ministere}: {total:,.0f} millions d'euros")
```

```python
# VISUALISATION 3: Analyse par postes budgétaires
print("=== VISUALISATION 3: ANALYSE PAR POSTES BUDGÉTAIRES ===")

# Calculer les totaux par poste
poste_totals = df_melted.groupby('Postes')['Balance'].sum().sort_values(ascending=False)
poste_totals_millions = poste_totals / 1e6

# Créer le graphique
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 12))

# Graphique 1: Top 15 des postes par balance totale
top_15_postes = poste_totals_millions.head(15)
colors = plt.cm.viridis(np.linspace(0, 1, len(top_15_postes)))

bars = ax1.barh(range(len(top_15_postes)), top_15_postes.values, color=colors)
ax1.set_yticks(range(len(top_15_postes)))
ax1.set_yticklabels([label[:40] + '...' if len(label) > 40 else label for label in top_15_postes.index])
ax1.set_xlabel('Balance Totale (Millions d\'euros)', fontsize=12)
ax1.set_title('Top 15 des Postes Budgétaires par Balance Totale (2012-2022)', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='x')

# Ajouter les valeurs sur les barres
for i, (bar, value) in enumerate(zip(bars, top_15_postes.values)):
    ax1.text(value + (max(top_15_postes.values) * 0.01), i, f'{value:,.0f}M€', 
             va='center', fontsize=8, fontweight='bold')

# Graphique 2: Heatmap des postes par année
# Sélectionner les 10 principaux postes
top_10_postes = poste_totals_millions.head(10).index
poste_evolution = df_melted[df_melted['Postes'].isin(top_10_postes)].groupby(['Année', 'Postes'])['Balance'].sum().unstack(fill_value=0)
poste_evolution_millions = poste_evolution / 1e6

# Créer la heatmap
im = ax2.imshow(poste_evolution_millions.T, cmap='RdBu_r', aspect='auto')
ax2.set_xticks(range(len(poste_evolution_millions.index)))
ax2.set_xticklabels(poste_evolution_millions.index)
ax2.set_yticks(range(len(poste_evolution_millions.columns)))
ax2.set_yticklabels([label[:30] + '...' if len(label) > 30 else label for label in poste_evolution_millions.columns])
ax2.set_title('Heatmap des 10 Principaux Postes par Année (Millions d\'euros)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Année', fontsize=12)
ax2.set_ylabel('Postes Budgétaires', fontsize=12)

# Ajouter une colorbar
cbar = plt.colorbar(im, ax=ax2)
cbar.set_label('Balance (Millions d\'euros)', fontsize=12)

plt.tight_layout()
plt.show()

# Statistiques détaillées
print(f"\nStatistiques par poste budgétaire :")
for poste, total in top_15_postes.items():
    print(f"  {poste}: {total:,.0f} millions d'euros")
```

```python
# ANALYSE 1: Identification des tendances et anomalies
print("=== ANALYSE 1: TENDANCES ET ANOMALIES ===")

# Calculer les variations annuelles
balance_totals_df = pd.DataFrame({
    'Année': balance_totals_millions.index,
    'Balance': balance_totals_millions.values
})
balance_totals_df['Variation_absolue'] = balance_totals_df['Balance'].diff()
balance_totals_df['Variation_pourcentage'] = balance_totals_df['Balance'].pct_change() * 100

print("Évolution des balances totales avec variations :")
print(balance_totals_df.round(2))

# Identifier les années avec les plus fortes variations
max_increase = balance_totals_df['Variation_absolue'].max()
max_decrease = balance_totals_df['Variation_absolue'].min()

print(f"\nPlus forte augmentation : {max_increase:,.0f} millions d'euros")
print(f"Plus forte diminution : {max_decrease:,.0f} millions d'euros")

# Analyser la volatilité
volatility = balance_totals_df['Balance'].std()
print(f"Volatilité (écart-type) : {volatility:,.0f} millions d'euros")

# Identifier les postes avec les plus fortes variations
print(f"\n=== ANALYSE DES VARIATIONS PAR POSTE ===")

# Calculer la variance par poste
poste_variance = df_melted.groupby('Postes')['Balance'].agg(['mean', 'std', 'min', 'max']).fillna(0)
poste_variance['range'] = poste_variance['max'] - poste_variance['min']
poste_variance = poste_variance.sort_values('range', ascending=False)

print("Top 10 des postes avec la plus forte variabilité :")
for i, (poste, row) in enumerate(poste_variance.head(10).iterrows()):
    print(f"  {i+1}. {poste}:")
    print(f"     Range: {row['range']:,.0f}€, Écart-type: {row['std']:,.0f}€")
```

```python
# ANALYSE 2: Corrélations et insights
print("=== ANALYSE 2: CORRÉLATIONS ET INSIGHTS ===")

# Créer une matrice de corrélation entre les années pour les principaux postes
top_10_postes_names = poste_totals_millions.head(10).index
correlation_data = df_melted[df_melted['Postes'].isin(top_10_postes_names)].groupby(['Année', 'Postes'])['Balance'].sum().unstack(fill_value=0)

# Calculer la matrice de corrélation
correlation_matrix = correlation_data.corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .8}, 
            xticklabels=[label[:20] + '...' if len(label) > 20 else label for label in correlation_matrix.columns],
            yticklabels=[label[:20] + '...' if len(label) > 20 else label for label in correlation_matrix.columns])
plt.title('Matrice de Corrélation entre les Principaux Postes Budgétaires', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Identifier les corrélations les plus fortes
print("Corrélations les plus fortes entre postes :")
corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:  # Seuil de corrélation forte
            corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_value))

corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
for poste1, poste2, corr in corr_pairs[:5]:
    print(f"  {poste1[:30]}... ↔ {poste2[:30]}... : {corr:.3f}")

# Analyser les postes avec des balances positives vs négatives
print(f"\n=== ANALYSE DES SIGNAUX BUDGÉTAIRES ===")

# Séparer les postes en recettes et dépenses
poste_means = df_melted.groupby('Postes')['Balance'].mean()
recettes = poste_means[poste_means > 0].sort_values(ascending=False)
depenses = poste_means[poste_means < 0].sort_values(ascending=True)

print(f"Top 5 des postes générant des recettes (moyenne positive) :")
for i, (poste, balance) in enumerate(recettes.head(5).items()):
    print(f"  {i+1}. {poste}: {balance:,.0f}€")

print(f"\nTop 5 des postes générant des dépenses (moyenne négative) :")
for i, (poste, balance) in enumerate(depenses.head(5).items()):
    print(f"  {i+1}. {poste}: {balance:,.0f}€")
```

**On remarque 2 choses selon les infos de notre data frame.**


*   Les "-" dans le fichier ne sont pas considérés comme étant des éléments nulls
*   Mise à part le numéro de compte, même les balances de sorties sont comptés comme étant des "object" et non des flottants
